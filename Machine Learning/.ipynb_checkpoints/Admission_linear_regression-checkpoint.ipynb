# Hey there, Welcome!
#### In this notebook we shall solve a problem on predicting the chance/probability of a candidate to get admission in XYZ college given his CGPA, TOFEL, GRE score and his Research work.

Importing all the required libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
from sklearn import linear_model
from sklearn.metrics import r2_score,mean_squared_error
```

Filepath of the dataset

```python
filepath = r"C:/Users/riaz/Desktop/Admission_Predict.csv"
df = pd.read_csv(filepath,index_col='Serial No.')
```

Checking for Null values

```python
df.isnull().sum()
```

Renaming columns and plotting thier Karl Pearson's Coefficients using corr() and heatmap

```python
df.columns=['gre','tofel','ranking','sop','lor','cgpa','research','admit']
sns.heatmap(df.corr(),annot=True)
```

```python
df.corr().admit
```

# Cleaning the dataset

Taking only the required feature that has high corr with the admit column

```python
new_df = df[['gre','cgpa','tofel','admit']]
```

```python
sns.heatmap(new_df.corr(),annot=True)
```

# Finding outliers

```python
def fourplots(listofplots,row,col,comment,data):
    fig, axs = plt.subplots(1, 4,figsize=(15,3))
    sub = []
    n = 0
    fig.suptitle(comment,fontsize=20,color='Gray')
    for i in listofplots:
        fig = sns.distplot(data[i],ax=axs[n])
        sub.append(axs)
        n+=1
```

Before removing outliers

```python
list_of_plots = ['gre','cgpa','tofel','admit']
row = 1
col = 3
comment = 'Before removing outliers'
fourplots(list_of_plots,row,col,comment,new_df)
```

```python
z = np.abs(stats.zscore(new_df))
```

After removing outliers

```python
clean = new_df[(z < 3).all(axis=1)]
comment = 'After removing outliers'
fourplots(list_of_plots,row,col,comment,clean)
```

# spliting X and Y terms

```python
X = clean.drop(['admit'],axis=1).values
y = clean['admit'].values
```

```python
y.shape
```

# Preprocessing the data 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

# Building the model

```python
reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_test.shape
```

Predicting the values of X_test

```python
pred = reg.predict(X_test)
```

Checking the score of predicted values

```python
reg.score(X_train,y_train)
```

# Visualising the predictions with true values

```python
plt.figure(figsize=(20,8))
y=pred
y1=y_test
x=np.arange(1, 133, 1)
x1=np.arange(0,133,10)
plt.plot(x,y,color='r',marker='o',label='Predicted',linestyle='dashed')
plt.plot(x,y1,color='g',label='Actual')
plt.xticks(x1)
plt.gca().legend(('Predicted','Test'))
plt.xlabel('Cases',fontsize=20)
plt.ylabel('Chance of getting admitted',fontsize=20)
plt.title('Predictions Vs True Values',fontsize=25)
plt.grid()
plt.ioff()
```

# Model Evaluations

```python
print("R2 score of our Model: ", end=" ")
print(r2_score(y_test,pred))
print("RMSE of our model: " ,end=' ')
print(np.sqrt(mean_squared_error(y_test,pred)))
```

Comparing the R2 score and Root Mean Square Error with our predictions, it seems that our model is pretty good to work with!
