```python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#raw_img = cv2.cvtColor(raw_img,cv2.COLOR_BGR2RGB)
raw_img = cv2.imread(r'C:/Users/riaz/Desktop/1.jpg')
crop = raw_img[200:450,600:800]
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(raw_img)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(crop)
```

```python
img = cv2.medianBlur(crop,3)
ret, img = cv2.threshold(img,130,255,type=cv2.THRESH_BINARY)
img = cv2.resize(img,(28,28),interpolation = cv2.INTER_NEAREST)
plt.imshow(img)
```

```python
image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(image,cmap='gray')


from tensorflow.keras.models import load_model
```

```python
model = load_model(r'C:\Users\riaz\Documents\GitHub\myPrograms\CNN models\MNIST\my_mnist_model.h5')
```

```python
#y_img = image/255
my_img = np.expand_dims(image,axis=0)
my_img = np.expand_dims(my_img, axis=-1)
```
